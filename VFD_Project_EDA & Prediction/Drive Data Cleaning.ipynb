```python
data['Stock'] = data['Stock'].str.replace('Currently out of Stock', 'Out of Stock')
data['Stock'] = data['Stock'].str.replace('In stock', 'In Stock')
data['Stock'] = data['Stock'].str.replace('2-4 weeks', '2-4 Weeks')
data['Stock'] = data['Stock'].str.replace('3-5 days', '3-5 Days')
```

```python
data['Model_Number'] = data['Description'].str.extract(r'([^:]+):\s*',expand=True)
data['Model_Number']
```

```python
data['Type'] = data['Description'].str.extract(r'[^:]+:\s*([^,]+,\s*[^,]+),\s*',expand=True)
data['Type']
```

```python
data['Motor_Type'] = data['Type'].str.extract(r'([^,]+)',expand=True)
data['Motor_Type'].unique()
```

```python
#data['Motor_Type'].str.extract(r'((?:Hitachi).*?(?:HP))',expand=True)
#data['Motor_Type'].str.extract(r'(Hitachi.*?HP)',expand=True)
data['Motor_Type'].apply(lambda text: re.sub(r'((Hitachi).*?(HP))', '', text) if isinstance(text, str) else text).str.strip()
#str.extract(r'(Hitachi.*?HP)',expand=True)
x = data['Motor_Type'].str.split('Hitachi')
data['Motor_Type'].unique()
```

```python
data['Motor_Type'] = data['Motor_Type'].apply(lambda text: re.sub(r'Hitachi\s*\d+/?\d*\s*HP', '', text) if isinstance(text, str) else text).str.strip()
data['Motor_Type'] = data['Motor_Type'].apply(lambda text: re.sub(r'Non Regenerative\s*\d+/?\d*\s*HP', 'Non Regenerative', text) if isinstance(text, str) else text)
data['Motor_Type'] = data['Motor_Type'].apply(lambda text: re.sub(r'(\d+\s?HP)', '', text) if isinstance(text, str) else text)
data['Motor_Type'] = data['Motor_Type'].apply(lambda text: re.sub(r'[0-9/.]', '', text) if isinstance(text, str) else text).str.strip()
data['Motor_Type'].unique()
```

```python
data['Drive_Type'] = data['Description'].str.extract(r',\s*(VFD)(?:,|$)*',expand=True)
data['Drive_Type']
```

```python
data['HP_rating'] = data['Description'].str.extract(r'(\d+\s?HP)',expand=True)
print(data['HP_rating'].unique())
data['HP_rating'] =  data['HP_rating'].str.replace('HP','').astype(float)
print(data['HP_rating'].unique())
```

```python
data['kW_rating'] = data['Description'].str.extract(r'(\d+\s?(?:kW|KW))',expand=True)
print(data['kW_rating'].unique())
data['kW_rating'] =  data['kW_rating'].str.replace('kW', '').str.replace('KW','').astype(float)
print(data['kW_rating'].unique())
```
