```python
#Build a prediction model to predict the price for a given HP_rating
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
x = new_drive_data[['HP_rating']] # Independent variable
y = new_drive_data['Price'] # Dependent variable
```

```python
#Split the training & testing sets
X_train, X_test = train_test_split(x, test_size=0.3, random_state=10, shuffle = True)
Y_train, y_test = train_test_split(y, test_size=0.3, random_state=10, shuffle = True)
model = LinearRegression()
model.fit(X_train, Y_train)
```

```python
y_predict = model.predict(X_test)
```

```python
y_predict
```

```python
plt.scatter(y_test, y_predict)
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.title('Actual vs Predicted Price')
plt.show()
```

```python
print(f'Intercept:{model.intercept_}')
print(f'Coefficient:{model.coef_}')
```

## Evaluate Model Performance
### There are three primary metrics used to evaluate linear models. These are: Mean absolute error (MAE), Mean squared error (MSE), or Root mean squared error (RMSE).

```python
from sklearn import metrics
import numpy as np
print(f'Model Mean absolute error (MAE) is {metrics.mean_absolute_error(y_test, y_predict)}')
print(f'Model Mean squared error (MSE) is {metrics.mean_squared_error(y_test, y_predict)}')
rmse = np.sqrt(metrics.mean_squared_error(y_test, y_predict))
print(f'Root mean squared error (RMSE) is {rmse}')
```

### Significant difference between RMSE and MAE value indicate that outliers are present in the data.
