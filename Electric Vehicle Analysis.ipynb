```python
#create a function to define the prices in ranges
def create_price_range(df, column='Base MSRP'):
    def categorize_price(price):
        if price == 0:
            return "Unknown"
        elif price < 40000:
            return "Low"
        elif price < 60000:
            return "Medium"
        else:
            return "High"
    df["Price_Range_Category"] = df[column].apply(categorize_price)
    return df
ev_df2 = create_price_range(ev_df1,column="Base MSRP")
```

```python
ev_df2.info()
```

```python
ev_df2['Electric Range'].value_counts()
```

```python
#create a distance ranges using numpy
def create_electric_range(df,column="Electric Range"):
    conditions = [df[column]==0, df[column]<150, df[column]<300]
    categories =["Unknown", "Short", "Medium"]
    df["Electric_Range_Category"] = np.select(conditions, categories, default = "Long")
    return df
ev_df3 = create_electric_range(ev_df2,column="Electric Range")
```
